### CMPE 256 Assignment 5 -  K Mean Cluster (Shopping Index & Income Index)  ###

#### Adding and Ploting the Dataset 

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn import metrics
```

```python
dataset = pd.read_csv('./Desktop/CMPE-256/Assignments/Assignment-5/Data_Points.csv')
dataset.head()
```

```python
import matplotlib.pyplot as plt

plt.plot()
plt.title('Data Points')
plt.xlabel('Spending Index')
plt.ylabel('Income Index')
plt.scatter(dataset['SpendingIndex'], dataset['IncomeIndex'])
```

#### K-Means with 2 Clusters

```python
k = 2
dataframe = dataset
kmeans_model = KMeans(n_clusters=k, random_state=0)
labels = kmeans_model.fit_predict(dataframe)
print(labels)
```

```python
dataframe.shape
```

```python
plt.plot()

filtered_label0 = dataframe[labels == 0]
 
filtered_label1 = dataframe[labels == 1]

#Plotting the results
plt.title('Clusters of Shoppers')
plt.xlabel('Spending Index')
plt.ylabel('Income Index')
plt.scatter(filtered_label0['SpendingIndex'], filtered_label0['IncomeIndex'], color = 'red', label = 0)
plt.scatter(filtered_label1['SpendingIndex'], filtered_label1['IncomeIndex'], color = 'white', label = 1)

plt.legend()
plt.show()
```

```python
#Getting the Centroids
from turtle import back


centroids = kmeans_model.cluster_centers_
u_labels = np.unique(labels)
 
#Plotting the results:
plt.title('Clusters of Shoppers')
plt.xlabel('Spending Index')
plt.ylabel('Income Index')

for i in u_labels:
    plt.scatter(dataframe[labels == i]['SpendingIndex'] , dataframe[labels == i]['IncomeIndex'] , label = i)
    
plt.scatter(centroids[:,1] , centroids[:,2] , s = 80, color = 'white')

plt.legend()
plt.show()
```
